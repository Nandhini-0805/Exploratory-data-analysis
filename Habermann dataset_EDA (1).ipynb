#OBJECTIVE

#the age of patients lie around 58-69
#BALANCED OR IMBALANCED?
#it is imbalanced.

```python
#PAIR PLOTS
sns.set_style('whitegrid')
sns.pairplot(ds,hue='status',size=3,diag_kind='kde')
pt.show()
```

#observation:from above pairplots we can observe woman can be diagnized with breast cancer between the age of 25-80
#the no of patients survived post surgery are greater than the no of patients dead is between the year 1957 to 1968


```python
#UNIVARIATE ANALYSIS:pdf
sns.FacetGrid(ds, hue="status",height=6) \
   .map(sns.distplot, "age") \
   .add_legend();
pt.xlabel('age')
pt.ylabel('status')
pt.show();
```

the observation for above pdf with age and status shows there is a lot of overlap in the plot.Since there is a strong overlap we cant come to conclusion

```python
#PDF WITH RESPECT TO YEAR
sns.FacetGrid(ds, hue="status",height=6) \
   .map(sns.distplot, "year") \
   .add_legend();
pt.xlabel('year')
pt.ylabel('status')
pt.show();
```

#Overlapping are visible.
#Between the year 1962 and 1966 we could see there are more no. of deaths 
#Between 1957 nd 1960 more no. of survivors post surgery are visible

```python
#box plots:
sns.boxplot(x='status',y='age',data=ds)
pt.show()
sns.boxplot(x='status',y='year',data=ds)
pt.show()
sns.boxplot(x='status',y='nodes',data=ds)
pt.show()
```

```python
sns.violinplot(x='status',y='age',data =ds,height = 7)
pt.show()
sns.violinplot(x='status',y='year',data =ds,height = 7)
pt.show()
```

#observatioN:Both Box and Violin plots shows high range of overlaps
#Operated patients around 1962-1965 are less likely to survive
#With age parameter considered, patients with age  46-67 did not survive 

```python
sns.jointplot(x = 'year', y = 'age', data = ds, kind = 'kde')
pt.grid()
pt.show()
```

#conclusion:
1.Age is definitely not a factor of diagnosis.
2.But patients who are less than 35 years of age are more likely to survive
3.Patients Patients who are greater than 75 shows no status of survival
4.As the given data is not sufficient to classify ,the status is IMBALANCED.
